```python
train_labels = pd.read_csv('../input/just-the-basics-the-after-party/train_labels.csv', header = None)
train = pd.read_csv('../input/just-the-basics-the-after-party/train.csv', header = None)
test = pd.read_csv('../input/just-the-basics-the-after-party/test.csv', header = None)
```

```python
train = train.fillna(train.mean())
test = test.fillna(test.mean())
```

```python
train.info()
```

```python
from sklearn.preprocessing import StandardScaler

std = StandardScaler()

x_std = std.fit_transform(train)
```

```python
from sklearn.model_selection import train_test_split


train_labels = np.ravel(train_labels)

x_train, x_test, y_train, y_test = train_test_split(x_std, train_labels, random_state = 42 )
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 7)

knn.fit(x_train,y_train)

knn.predict(x_test)

knn.score(x_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
neighbors = np.arange(1,50)

train_acc = []

val_acc = []

kfold = 10

best_knn = None

best_acc = 0.0

for i, k in enumerate(neighbors):
    
    
    knn = KNeighborsClassifier(n_neighbors = k)
    
    knn.fit(x_train, y_train)
    
    train_acc.append(knn.score(x_train, y_train))
    
    nominee = np.mean(cross_val_score(knn, x_std, train_labels, cv = kfold))
    
    val_acc.append(nominee)
    
    if nominee > best_acc:
        
        best_acc = nominee
        best_knn = knn
    
```

```python
print(best_acc)
print(knn)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()

rfc.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
```

```python
ac = accuracy_score(y_test, rfc.predict(x_test))

f1 = f1_score(y_test, rfc.predict(x_test))
```
