```python
import numpy as np
import pandas as pd
```

```python
train=pd.read_csv('../input/ventilator-pressure-prediction/train.csv')
test=pd.read_csv('../input/ventilator-pressure-prediction/test.csv')
```

```python
train
```

```python
train.corr()['pressure']
```

```python
train=train.drop(['breath_id'],axis=1)
test=test.drop(['breath_id'],axis=1)
```

```python
train.info()
```

```python
train=train.drop(['id'],axis=1)
test=test.drop(['id'],axis=1)
```

```python
X=train.drop('pressure',axis=1)
y=train['pressure']
X.shape
train.shape
```

```python
test.shape
```

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model=keras.Sequential([
    layers.Dense(512,activation='relu',input_shape=[5]),
    layers.Dense(256,activation='relu'),
    layers.Dense(128,activation='relu'),
    layers.Dense(64,activation='relu')
])
```

```python
model.compile(optimizer='adam',loss='mae')
```

```python
history=model.fit(X,y,validation_data=(X,y),batch_size=256,epochs=10)
```

```python
y_preds=model.predict(test)
```

```python
submission.=pd.read_csv('../input/ventilator-pressure-prediction/sample_submission.csv')
```

```python
submission['pressure']=y_preds
```

```python
submission.to_csv('submisson.csv')
```
