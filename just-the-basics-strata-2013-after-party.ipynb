```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('../input/just-the-basics-the-after-party/train.csv')
test = pd.read_csv('../input/just-the-basics-the-after-party/test.csv')
y = pd.read_csv('../input/just-the-basics-the-after-party/train_labels.csv')
```

```python
train.head()
```

```python
col = list(range(0,100))
train.columns = [col]
test.columns = [col]
```

```python
train.head()
```

```python
train.info()
```

```python
for i in col:
    train[i,].fillna(train[i,].median(), inplace=True)
for i in col:
    test[i,].fillna(test[i,].median(),inplace=True)
```

```python
train.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for column in train:
        plt.figure(figsize=(10,1))
        sns.boxplot(data=train, x=column)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import StackingClassifier
```

```python
estimators = [
('rf', RandomForestClassifier(n_estimators=10, random_state=42)),
('svr', make_pipeline(StandardScaler(),
 LinearSVC(random_state=42)))
]
clf = StackingClassifier(
     estimators=estimators, final_estimator=LogisticRegression()
 )
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    train, y, stratify=y, random_state=42
 )
clf.fit(X_train, y_train).score(X_test, y_test)
```

```python
ans=clf.predict(test)

f=open("/kaggle/working/answers.csv", "w")
f.write(str(ans))
f.close()
```
